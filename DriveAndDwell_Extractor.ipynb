# **EKAR PLUS**

```python
import requests
from bs4 import BeautifulSoup

url = "https://ekarplus.com/"
url1 = "https://ekarplus.com/realEstate-card.xhtml?id=39812"

payload = {}
headers = {
    'Cookie': 'JSESSIONID=jpkAb634lhf54wsCayIt64eA.vmi1396406'
}

response = requests.request("GET", url, headers=headers, data=payload)

headers['Cookie'] = '; '.join([f"{cookie.name}={cookie.value}" for cookie in response.cookies])

response1 = requests.request("GET", url1, headers=headers, data=payload)
soup = BeautifulSoup(response1.content, 'html.parser')
col_md_8 = soup.find(class_='col-md-8')

col_md_6 = soup.find(class_='col-md-6')
val2= col_md_6.get_text(strip=True, separator=': ').split('|')

col_md_3 = soup.find(class_='col-md-3')
col_md_6 = col_md_3.find_next(class_='col-md-6')
text_content = col_md_6.get_text(strip=True)
val3 = [part.strip(': ') for part in text_content.split(': ')]

small_elements = soup.find_all('small', class_='col-md-3')

diction1 = {
    'Description': soup.find(class_='form-control').get_text(strip=True, separator=' ').split('|'),
    'Location': val2,
    'Person_Info': val3
}

for small in small_elements:
    label = small.find('label').get_text(strip=True)
    value = small.get_text(strip=True, separator=' ').replace(label, '').strip()
    diction1[label] = value

for key, value in diction1.items():
    print(f"{key}: {value}")
```
